# MIR은 무엇인가?

MIR(music information retrieval)

한국어로는 "음악정보검색"

------
MIR의 과제들을 살펴보자

- 오디오 지문 (fingerprinting)
- 커버 송 감지 (cover song detection)
- 장르 인지 (genre recognition)
- 전사 (transcription)
- 추천 (recommendation)
- 심볼릭 멜로디 유사도 (symbolic melodic similarity)
- 분위기 (mood)
- 소스 분류 (source separation)
- 악기 인지 (instrument recognition)
- 피치 트랙킹 (pitch tracking)
- 템포 추정 (tempo estimation)
- 악보 정렬 (score alignment)
- 노래 구조 (song structure/form)
- 비트 트래킹 (beat tracking)
- 키 감지 (key detection)
- 허밍 기반 검색 (query by humming)
